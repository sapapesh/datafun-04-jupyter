### *Project Step 1* - Import the Data


The following code import the libraries needed for the project.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### *Project Step 2* - Load the Data

The following code imports the Iris dataset and then shows the first 5 columns for the data.

```python
# Load the Iris dataset into DataFrame
df = sns.load_dataset('iris')

# Inspect first rows of the DataFrame
print(df.head())
```

### *Project Step 3* - Inspecting the data

The following code will display the first 10 rows of the dataframe.  Then it will give the shape of the data by presenting the number of rows and columns (rows, columns).  Finally, it will tell us the data type of each column.

```python
print(df.head(10))
print(df.shape)
print(df.dtypes)
```

### *Project Step 4* - Descriptive statistics

The following code will display the summary statistics of each column containing numeric data.

```python
print(df.describe())
```

### *Project Step 4* - Data distribution for numerical columns

The following code will display a histograms of the different species, sepal length, sepal width, petal length, and petal width.

```python
# Inspect histogram by numerical column
df['petal_length'].hist()

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

**Analysis**

### *Project Step 5* - Data Distribution by species

The following code will display a histograms of the different species.

```python
# Inspect value counts by categorical column
print(df['species'].value_counts())

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
```

The data appears to be equally distributed among the different species of iris.

### *Project Step 6* - Data Transformation

The following code will display a histograms of the sepal area and the petal area.

```python
#Adding new columns
df['sepal_area'] = df['sepal_length'] * df['sepal_width']

df['petal_area'] = df['petal_length'] * df['petal_width']

# Inspect histograms for all numerical columns
df.hist()

# Show all plots
plt.show()
```

**Analysis**

* Sepal area
    1. There is a small subset with an area up to 30 cm.
    2. The median of the area is smaller than I would have expected based on the sepal length and sepal width charts.

* Petal area
    1. There is a large number of iris with petals with an area of 1 Cm.
    2. The distribution of the remaining petal area results is similar to what was expected based on the petal length and petal area charts.

*In looking at the trends, I am wondering if the species has a greater effect on the petal width and length than on the sepal width and length.*

### ***Conclusion***

The versicolor iris appears to be fairly average for all characteristics being reviewed. It is on the smaller side for the sepal width and, consequently, the sepal area.
